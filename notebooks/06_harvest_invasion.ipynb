```python
from pathlib import Path
import subprocess
import pandas as pd

ROOT = Path('/Users/ameerfiras/invasion-commons/invasion-commons')
RUNS_PREFIX = ROOT / 'results/runs/harvest_invasion/curated/harvest_invasion_stageB_pilot_v2'
SUMMARY_PREFIX = ROOT / 'results/runs/showcase/curated/harvest_invasion_stageB_pilot_v2'

RUNS_CSV = RUNS_PREFIX.with_name(RUNS_PREFIX.name + '_runs.csv')
HISTORY_CSV = RUNS_PREFIX.with_name(RUNS_PREFIX.name + '_generation_history.csv')
STRATEGY_CSV = RUNS_PREFIX.with_name(RUNS_PREFIX.name + '_strategy_history.csv')
TABLE_CSV = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_table.csv')
RANKING_CSV = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_ranking.csv')
DELTA_CSV = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_delta.csv')
CI_CSV = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_ci.csv')
SUMMARY_MD = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_summary.md')
MAIN_FIG = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_main.png')
MECH_FIG = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_mechanisms.png')
```

## Regenerate the summary and figures

```python
if RUNS_CSV.exists():
    subprocess.run([
        'python', '-m', 'experiments.summarize_harvest_invasion',
        '--runs-csv', str(RUNS_CSV),
        '--output-prefix', str(SUMMARY_PREFIX),
    ], cwd=ROOT, check=True)
    subprocess.run([
        'python', '-m', 'experiments.plot_harvest_invasion',
        '--ci-csv', str(CI_CSV),
        '--output-prefix', str(SUMMARY_PREFIX),
    ], cwd=ROOT, check=True)
else:
    print('Run the matrix first so the runs CSV exists.')
```
