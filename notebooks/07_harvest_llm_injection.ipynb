This notebook is the reproducibility companion for the Harvest Commons live LLM injection cycle. It is wired to the Stage A smoke run first, and can be pointed at the Stage B or Stage C outputs once those finish.

## What this notebook covers

1. Point to a Harvest LLM run prefix.
2. Regenerate the summary, ranking, paired deltas, confidence intervals, and figures.
3. Inspect LLM validity and fallback rates.
4. Read the paired `hybrid - top_down_only` deltas without manual CSV editing.

```python
from pathlib import Path
import subprocess
import pandas as pd

ROOT = Path('/Users/ameerfiras/invasion-commons/invasion-commons')
RUNS_PREFIX = ROOT / 'results/runs/harvest_invasion/curated/harvest_invasion_llm_stageA_smoke'
SUMMARY_PREFIX = ROOT / 'results/runs/showcase/curated/harvest_invasion_llm_stageA_smoke'

# Switch these two lines later if you want Stage B or Stage C instead.
# RUNS_PREFIX = ROOT / 'results/runs/harvest_invasion/curated/harvest_invasion_llm_stageB'
# SUMMARY_PREFIX = ROOT / 'results/runs/showcase/curated/harvest_invasion_llm_stageB'

RUNS_CSV = RUNS_PREFIX.with_name(RUNS_PREFIX.name + '_runs.csv')
TABLE_CSV = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_table.csv')
RANKING_CSV = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_ranking.csv')
DELTA_CSV = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_delta.csv')
CI_CSV = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_ci.csv')
INTEGRITY_CSV = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_integrity.csv')
SUMMARY_MD = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_summary.md')
COMPACT_FIG = SUMMARY_PREFIX.with_name(SUMMARY_PREFIX.name + '_compact.png')
RUNS_CSV.exists(), SUMMARY_MD.exists()
```

## Regenerate the Harvest LLM summaries and figures

```python
if RUNS_CSV.exists():
    subprocess.run([
        'python', '-m', 'experiments.summarize_harvest_invasion',
        '--runs-csv', str(RUNS_CSV),
        '--output-prefix', str(SUMMARY_PREFIX),
    ], cwd=ROOT, check=True)
    subprocess.run([
        'python', '-m', 'experiments.plot_harvest_invasion',
        '--ci-csv', str(CI_CSV),
        '--output-prefix', str(SUMMARY_PREFIX),
    ], cwd=ROOT, check=True)
else:
    print('Run the Harvest LLM matrix first so the runs CSV exists.')
```
