# 01 — Fishery Basics

## 1. Environment Definition

We define a minimal global-stock fishery with:

- $N$ agents  
- Global stock $S_t \in [0, S_{\max}]$  
- Individual harvest $h_i^t \in [0, h_{\max}]$

### Stock Update Dynamics

After harvest:

$$
S_t^{\text{remaining}} = 
\max\left(0, S_t - \sum_{i=1}^{N} h_i^t \right)
$$

Regrowth follows a logistic-like dynamic:

$$
G(S) = r \, S \left(1 - \frac{S}{S_{\max}}\right)
$$

Stock update:

$$
S_{t+1} =
\min\left(
S_{\max},
S_t^{\text{remaining}} + G(S_t^{\text{remaining}})
\right)
$$

Where:

- $r$ = regeneration rate  
- $S_{\max}$ = carrying capacity  


---

## 2. Collapse Rule

Define collapse threshold $S_c$.

If:

$$
S_t < S_c
\quad \text{for } K \text{ consecutive steps}
$$

Then:

$$
\text{Collapse} \Rightarrow
S_t = 0
\quad \forall t' > t
$$

This introduces path dependence.


---

## 3. Observation Model

Agents do not necessarily observe true stock.

Observed stock:

$$
\tilde{S}_t = S_t + \epsilon_t
$$

Where:

$$
\epsilon_t \sim \mathcal{N}(0, \sigma^2)
$$

Clipped to valid range:

$$
\tilde{S}_t \in [0, S_{\max}]
$$

This introduces epistemic uncertainty.


---
## 4. Strategy Composition

Population consists of:

- $n_g$ greedy agents  
- Remaining agents split among:
  - Conditional cooperators  
  - Conservative agents  
  - Punishers  

Total agents fixed:

$$
N = 8
$$

We sweep:

$$
n_g \in \{0, 1, \dots, 8\}
$$


---
## 5. Empirical Quantity Measured

For each configuration:

$$
\text{CollapseRate}(n_g, r, \sigma)
=
\frac{\text{Number of collapsed runs}}{\text{Total seeds}}
$$

Where:

- seeds = 200  
- Each seed initializes RNG independently  


---
## 6. Empirical Findings So Far

For $r = 1.5$:

**Without noise:**

- Sharp deterministic tipping point at $n_g = 3$

**With noise:**

- Smooth probabilistic transition region  
- Collapse probability increases continuously near threshold  

This indicates:

The system exhibits **stochastic tipping dynamics**.


---
## 7. Interpretation

The system behaves like a noisy nonlinear dynamical system with a critical composition threshold.

Let:

$$
T(r)
$$

be the minimal number of greedy agents that induces collapse.

Empirically:

- Below $T(r)$: stable regime  
- Near $T(r)$: probabilistic collapse  
- Above $T(r)$: deterministic collapse  

Observation noise widens the transition region.
